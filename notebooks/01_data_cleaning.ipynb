```python
import pandas as pd

df = pd.read_csv("../data/zomato.csv")

df.head(5)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df = df.drop_duplicates()
```

```python
import numpy as np
df = df.replace("", np.nan)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=[
    "name",
    "area",
    "cuisines",
    "dinner_ratings",
    "delivery_ratings",
    "averagecost"
])
```

```python
df["delivery_ratings"] = pd.to_numeric(df["delivery_ratings"], errors="coerce")
df["dinner_ratings"] = pd.to_numeric(df["dinner_ratings"], errors="coerce")
```

```python
df = df[(df["delivery_ratings"] <= 5) & (df["delivery_ratings"] >= 0)]
df = df[(df["dinner_ratings"] <= 5) & (df["dinner_ratings"] >= 0)]
```

```python
df["cuisines"] = df["cuisines"].str.lower().str.strip()
```

```python
df = df[df["averagecost"] < df["averagecost"].quantile(0.99)]
```

```python
df["dinner_ratings"] = pd.to_numeric(df["dinner_ratings"], errors="coerce")
df["delivery_ratings"] = pd.to_numeric(df["delivery_ratings"], errors="coerce")
```

```python
df["dinner_reviews"] = pd.to_numeric(df["dinner_reviews"], errors="coerce")
df["delivery_reviews"] = pd.to_numeric(df["delivery_reviews"], errors="coerce")
```

```python
df["averagecost"] = pd.to_numeric(df["averagecost"], errors="coerce")
```

```python
if "timing" in df.columns:
    df["timing"] = df["timing"].str.replace("â€“", "-", regex=False)
```

```python
binary_cols = [
    "ishomedelivery",
    "istakeaway",
    "isindoorseating",
    "isvegonly"
]
```

```python
for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)
```

```python
df["city"] = "Bangalore"
```

```python
df["latitude"] = np.nan
df["longitude"] = np.nan
```

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
print("Cleaned Shape:", df.shape)
print(df.isnull().sum())
```

```python
df.to_csv("../data/zomato_cleaned.csv", index=False)

print("Data cleaning completed. File saved as zomato_cleaned.csv")
```
