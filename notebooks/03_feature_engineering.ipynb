```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/zomato_cleaned.csv")
```

```python
df["overall_rating"] = (df["dinner_ratings"] + df["delivery_ratings"]) / 2
```

```python
df["total_reviews"] = df["dinner_reviews"] + df["delivery_reviews"]
```

```python
df["demand_score"] = (
    0.6 * df["total_reviews"] +
    0.4 * df["overall_rating"] * 100
)
```

```python
df["cost_index"] = df["averagecost"] / df["averagecost"].mean()
```

```python
df["cuisine_count"] = df["cuisines"].apply(lambda x: len(str(x).split(",")))
```

```python
area_demand = df.groupby("area")["demand_score"].mean()

df["area_demand_score"] = df["area"].map(area_demand)
```

```python
area_density = df.groupby("area")["name"].count()
df["restaurant_density"] = df["area"].map(area_density)
```

```python
df["opportunity_score"] = df["demand_score"] / df["restaurant_density"]
```

```python
area_rating = df.groupby("area")["delivery_ratings"].mean()
df["area_avg_rating"] = df["area"].map(area_rating)
```

```python
df.to_csv("../data/zomato_feature_engineered.csv", index=False)

print("Feature engineering completed")
```
